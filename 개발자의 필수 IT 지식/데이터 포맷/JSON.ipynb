*JSON(JavaScript Object Notation)*

>JSON은 자바스크립트 객체 문법을 따르는 문자열로 JSON 안에는 마찬가지로 자바스크립의 undefined를 제외한 기본 데이터 타입인 문자열, 숫자, 배열, 불리언 그리고 다른 객체를 포함할 수 있습니다.





```python
"""
name이 sehyun이다. 라는 것을 key : value 형태로 나타내는 것이고,
중간에 콜론을 통해서 말입니다.!
{}중괄호로 감싸기 때문이지요
"""

{
    "name": "sehyun",
    "like": {
        "운동" :{
            "weight" : "10kg",
            "feature" : "육각형"
        },
        "바나나" : {
            "color":"노란색"
        }
    }
}
```

```python
"""
지브리 OST list

- 마녀 배달부 키키(따스함에 둘러쌓인다면)
- 하울의 움직이는 성(세계의 약속)

작품명(name-key)은 마녀 배달부 키키, 노래 제목(song-key)은 따스함에 둘러 쌓인다면 입니다.
"""

"""
앞의 코드처럼 표현이 가능하고, 배열은 []로 감싸서 표현하므로 JSON도 동일하고,
여러 개의 지브리 OST가 있기 때문에 해당 부분을 대괄호 기반으로 표현하고,
"""

{
    "지브리OST리스트" : [
        {
            "name" : "마녀 배달부 키키",
            "song" : "따스함에 둘러쌓인다면"
        },
        {
            "name" : "하울의 움직이는  성",
            "song" : "세계의 약속"
        }
    ]
}
```

```python
"""
각 객체가 다른 타입을 가지고 있어도 됩니다.
"""

{
    "지브리OST리스트":[
        {
            "name":"마녀 배달부 키키",
            "song":"띠스함에 둘러쌓인다면"  
        },
        {
            "name":"하울의 움직이는 성",
            "song":1
        }
    ]
}
```

```python
"""
자바스크립트와의 호환성이 좋다.
아래의 코드처럼 문자열로 받은 JSON 타입의 데이터 포맷을 JSON.parse 만을 통해 쉽게 
자바스크립트 객체(Object)로 변환하는 것을 볼 수 있다.
"""

let a ='
{
    "지브리OST리스트":[
        {
            "name":"마녀 배달부 키키",
            "song":"띠스함에 둘러쌓인다면"  
        },
        {
            "name":"하울의 움직이는 성",
            "song":1
        }
    ]
}'
a = JSON.parse(a)
console.log(a)
```

```python
"""
배열은 [0], [1] 이런 식으로 접근하며 되고
해당 key에 대한 value는 .Key 또는 ['key'] 이런 식으로 접근해서 빼내면 됩니다.
"""

const a = {
    "지브리OST리스트":[
        {
            "name":"마녀 배달부 키키",
            "song":"띠스함에 둘러쌓인다면"  
        },
        {
            "name":"하울의 움직이는 성",
            "song":"섹의 약속"
        }
    ]
}

console.log(a.지브리OST리스트[0])
console.log(a.지브리OST리스트[0].name)
console.log(a.지브리OST리스트[0]["song"])
```

## JSON 주의할 점

>JSON은 순수한 데이터 포맷으로 오직 key-value 들만 담을 수 있고, 메서드는 담을 수 없습니다. 그리고 작은 따옴표가 아닌 큰 따옴표만을 사용해야하고, undefined는 불가합니다.



```python
{
    "지브리OST리스트":[
        {
            'name':undefined,
            "song":"띠스함에 둘러쌓인다면"  
        },
        {
            "name":"하울의 움직이는 성",
            "song":"섹의 약속"
        }
    ]
}
```

### JSON의 자료형

자바스크립트와 유사하지만 undefined, 메서드 등을 포함하지 않습니다.

- 수(Number)
- 문자열(string)
- 참/거짓(Boolean)
- 배열(array)
- 객체(Object)
- null

<br>

### JSON의 장점

>JSON은 텍스트로 이루어져 있으므로, 사람과 컴퓨터 모두 읽고 쓰기 쉽고 프로그래밍 언어와 플랫폼에 독립적이므로, 서로 다른 시스템간에 객체를 교환하기 쉽기에 주로 API, config 파일에 활용되며 가볍습니다. 독립적인 특성으로 인해서 프런트언어와 백엔드 언어가 다름에도 서로 데이터 통신들이 가능합니다.
